```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return x*x+5
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
x1=np.linspace(-1,1)
x2=np.linspace(-6,6)
x3=np.linspace(0,5)
plt.subplot(1,3,1)
plt.plot(x1,f(x1))
plt.subplot(1,3,2)
plt.plot(x2,f(x2))
plt.subplot(1,3,3)
plt.plot(x3,f(x3))
plt.show()
```
