zadanie 3

zdefiniowanie wymaganej funkcji f(x)

```python
def f(x):
    return x*x+5
```

zdefiniowanie funkcji rysującej wartości funkcji dla zadanego przedziału

```python
def plot(mini, maxy):
    x=np.linspace(mini,maxy)
    plt.plot(x,f(x))
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('f(x) dla x z przedziału ('+str(mini)+','+str(maxy)+')')
    plt.legend('f(x)')
```

narysowanie wymaganych wykresów

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plot(-1,1)
plt.subplot(1,3,2)
plot(-6,6)
plt.subplot(1,3,3)
plot(0,5)
plt.show()
```
